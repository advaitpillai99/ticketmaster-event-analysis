```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\advpi\OneDrive\Desktop\Personal Projects\ticketmaster_events_2025_detailed.csv')
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
df.drop(columns=['Venue Capacity'], inplace=True)
```

```python
df['State'].fillna("Unknown", inplace=True)
```

```python
df['Promoter'].fillna("Not Listed", inplace=True)
```

```python
df.drop(columns=['Accessibility Info'], inplace=True)
```

```python
df.head()
```

```python
# Imputing the missing values in Min Price and Max Price using the median (as it is least affected by outliers)
df['Min Price'].fillna(df['Min Price'].median(), inplace=True)
df['Max Price'].fillna(df['Max Price'].median(), inplace=True)
print(df[['Min Price', 'Max Price']].isnull().sum())
```

```python
df.to_csv("C:/Users/advpi/OneDrive/Desktop/Personal Projects/ticketmaster_events_2025_cleaned.csv", index=False)
print("ðŸŽ‰ Cleaned dataset saved successfully!")
```
